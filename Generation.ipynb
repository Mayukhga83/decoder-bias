```python
import torch
torch.cuda.set_device(1)



torch.cuda.current_device()

################------import for Hugging Face Models-------##########

import json
import torch
from happytransformer import HappyGeneration
from happytransformer import GENSettings
import numpy as np

#happy_gen = HappyGeneration("Model name", "model name from transformers")
happy_gen = HappyGeneration("GPT2_Neo1.3B", "EleutherAI/gpt-neo-1.3B")



################------Functions for Generations-------##########


def test_T_forP(group,  num_generation, iterator = np.round(np.arange(0.2, 1, 0.1), 2).tolist(), top_p = 0.9, min_length = 45, max_length = 50 ):
    ''' TESTING T FOR FIXED VALUE OF P
        P = 0.9 by default, range (0, 1)
        T = iterator
    
    '''
    output = {}
    for prompts in group:
        data = {}
        
        for item in iterator:
            lis = []
            sampling_parameter = GENSettings(do_sample=True, no_repeat_ngram_size=2, early_stopping=False, top_p= top_p, temperature=item , min_length = min_length, max_length = max_length)
            for i in range(num_generation):
                result = happy_gen.generate_text(prompts, args = sampling_parameter)
                lis.append(result.text)
            data[str(item)] = lis
            
        output[prompts] = data
    return output


def test_T_forK(group, num_generation, iterator = np.round(np.arange(0.2, 1, 0.1), 2).tolist(), top_k = 50, min_length = 45, max_length = 50 ):
    ''' TESTING T FOR FIXED VALUE OF K
        K = 50 by default
        T = iterator
    
    '''
    output = {}
    for prompts in group:
        data = {}
        
        for item in iterator:
            lis = []
            sampling_parameter = GENSettings(do_sample=True, no_repeat_ngram_size=2, early_stopping=False, top_k= top_k, temperature=item, min_length = min_length, max_length = max_length)
            for i in range(num_generation):
                result = happy_gen.generate_text(prompts, args = sampling_parameter)
                lis.append(result.text)
            data[str(item)] = lis
            
        output[prompts] = data
    return output


def test_P_forT(group, num_generation, iterator = np.round(np.arange(0.2, 1, 0.1), 2).tolist(), temperature = 0.7, min_length = 45, max_length = 50 ):
    ''' TESTING P FOR FIXED VALUE OF T
        T = 0.7 by defaault
        P = iterator
    
    '''
    output = {}
    for prompts in group:
        data = {}
        
        for item in iterator:
            lis = []
            sampling_parameter = GENSettings(do_sample=True, no_repeat_ngram_size=2, early_stopping=False, top_p = item, temperature=temperature, min_length = min_length, max_length = max_length)
            for i in range(num_generation):
                result = happy_gen.generate_text(prompts, args = sampling_parameter)
                lis.append(result.text)
            data[str(item)] = lis
            
        output[prompts] = data
    return output

def test_K_forT(group, num_generation, iterator = np.round(np.arange(10, 120, 20), 2).tolist(), temperature = 0.7, min_length = 45, max_length = 50 ):
    ''' TESTING K FOR FIXED VALUE OF T
        T = 0.7 by defaault
        K = iterator  : a differenet iterator 
    
    '''
    output = {}
    for prompts in group:
        data = {}
        
        for item in iterator:
            lis = []
            sampling_parameter = GENSettings(do_sample=True, no_repeat_ngram_size=2, early_stopping=False, top_k = item, temperature=temperature, min_length = min_length, max_length = max_length)
            for i in range(num_generation):
                result = happy_gen.generate_text(prompts, args = sampling_parameter)
                lis.append(result.text)
            data[str(item)] = lis
            
        output[prompts] = data
    return output


prompt = ['Add the prompts here as a list of strings']
```
